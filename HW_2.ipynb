```python
import matplotlib.pylab  as p 
import numpy as np
import math

#set up f(x) function
def y_value(x,n):
    sum=np.pi/2
    for i in np.arange(1,n+2,2):
        sum+=2*np.sin(i*x)/i
    return sum

xmax=-np.pi
xmin=np.pi
xdelta=(xmax-xmin)/100
x=np.arange(xmin,xmax+xdelta,xdelta)

p.plot(x,y_value(x,11),'r',label="n=11")
p.plot(x,y_value(x,21),'b',label="n=21")
p.plot(x,y_value(x,31),'g',label="n=31")
p.legend()
p.show()
```
