```python
import numpy as np
import matplotlib.pyplot as plt

#forward difference
def derifw(y,x,dx):
    return (y(x+dx)-y(x))/dx

#central difference
def dericd(y,x,dx):
    return(y(x+dx/2)-y(x-dx/2))/dx

#set up x and dx
x=np.pi/4
dx=[]
for  i in range(1,17):
    dx.append(10**(-i))

#now we know y function is  f(x)=cos(x), so f'(x)=-sinx
df =  lambda x: -np.sin(x)

#Plot log(Error)=log10|f’(numerically)-f’(analytic)| against log10(dx) for dx=1e-16,1e-15,….,0.1 at x=π/4.
abs_error_fw=[]
abs_error_cd=[]
for i in range(len(dx)):
    abs_error_fw.append(abs(derifw(np.cos,x,dx[i])-df(x)))#forwar difference error
    abs_error_cd.append(abs(dericd(np.cos,x,dx[i])-df(x)))#central difference error

#start plot
plt.plot(dx,abs_error_fw,label='fw')
plt.plot(dx,abs_error_cd,label='cd')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('dx')
plt.ylabel('Error')
plt.title('Numerical Evaluation of the Derivative of cos(x), x=π/4')
plt.legend()
plt.show()
```

For fw,Error~dx^n,n=-8
For cd,Error~dx^n,n=-5
The central difference method also has better numerical stability and accuracy because the error in the derivative calculation decreases as h decreases.
